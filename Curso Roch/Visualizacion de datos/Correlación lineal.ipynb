```python
#Pruebo normalidad en un conjunto de datos
import numpy as np
data_points = np.random.normal(0, 1, 100)
data_points
```

```python
#Creo el histograma con los datos generados aleatoriamente
plt.hist(data_points, edgecolor="black", linewidth=2)
```

```python
#Creo el gráfico Quantile-Quantile para corroborar la normalidad
import pylab
import scipy.stats as stats
stats.probplot(data_points, dist="norm", plot=pylab)
pylab.show()
```

```python
#Aplico el test de Shapiro-Wilks para probar normalidad
from scipy.stats import shapiro
estadistico, p_value = shapiro(data_points)
print("Estadistico=%.3f, p_value=%.3f" % (estadistico, p_value))
```

```python
#Pruebo normalidad en algunas de las variables para determinar el tipo de coeficiente a utilizar para calcular la correlación
plt.hist(contamina["Velocidad_viento"], edgecolor="black",linewidth=1)
```

```python
#Creo el grafico Quantile-Quantile para Velocidad_Viento
stats.probplot(contamina["Velocidad_viento"], dist="norm", plot=pylab)
pylab.show()
```

```python
#Aplico el test de Shapiro-Wilks a Velocidad_Viento
estadistico, p_value = shapiro(contamina["Velocidad_viento"])
print("Estadistico=%.3f, p_value=%.3f" % (estadistico, p_value))
```

```python
#Creo histograma con la variable Contaminacion_SO2
plt.hist(contamina["Contaminacion_SO2"], edgecolor = "black", linewidth=2)
```

```python
#Creo el grafico Quantile-Quantile
stats.probplot(contamina["Contaminacion_SO2"], dist="norm", plot=pylab)
pylab.show()
```

```python
#Aplico el test de Shapiro-Wilks a Contaminacion_SO2
estadistico, p_value = shapiro(contamina["Contaminacion_SO2"])
print("Estadistico=%.3f, p_value=%.3f" % (estadistico, p_value))
```

```python
#Obtengo la correlación de todos los pares de variables
contamina_corr = contamina.corr(method="spearman")
contamina_corr
```

```python
#Creo mapa de calor con las correlaciones
sns.heatmap(contamina_corr,
           xticklabels=contamina_corr.columns,
           yticklabels=contamina_corr.columns,
           cmap="coolwarm"
           )
```

```python
#Para obtener la significancia uso pingouin
import pingouin as pg

corr = pg.pairwise_corr(contamina, method="spearman")

#Ordeno los valores de la correlacion en base a su significancia y le indico las columnas a desplegar
corr.sort_values(by=["p-unc"])[["X","Y","n","r","p-unc"]]
```
