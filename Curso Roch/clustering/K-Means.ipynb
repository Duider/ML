```python
# Es un método de clustering que puede analizar bases de datos con más de 10000 individuos. Para llevarlo a cabo es necesario 
# saber de antemano el número de clusters en los que vamos a dividir los datos.
# Cuando no se sabe la cantidad exacta de clusters se utiliza una tecnica llamada "codo de Jambu"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from warnings import simplefilter

%matplotlib inline
```

```python
vinos = pd.read_csv("../arch/caracteristicas de vinos.csv", engine="python")
vinos.info()
```

```python
vinos
```

```python
vinos_variables = vinos.drop(["Vino"], axis=1)
vinos_variables
```

```python
vinos_variables.describe()
```

```python
# Normalización de los valores
vinos_norm = (vinos_variables-vinos_variables.min())/(vinos_variables.max()- vinos_variables.min())
vinos_norm
```

```python
# Buscando la cantidad optima de clusters. Esto lo vamos a hacer calculando que tan similares son los individuos dentro de los cluster
#within clusters sum of squares?
simplefilter(action="ignore" , category=FutureWarning)
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, max_iter=300)
    kmeans.fit(vinos_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Grafico los resultados de WCSS para formar el Codo de Jambú
plt.plot(range(1,11), wcss)
plt.title("Codo de Jambu")
plt.xlabel("Numero de Clusters")
plt.ylabel("WCSS") #WCSS es un indicador de que tan similares son los individuos dentro de los clusters
plt.show()
```

```python
# Como podemos ver en el grafico el valor de WCSS dejo de disminuir de manera drastica al separar en tres clusters
```

```python
clustering = KMeans(n_clusters= 3, max_iter= 300) #Crea el modelo
clustering.fit(vinos_norm) # Aplica el modelo a la base de datos
```

```python
# Agrago la clasificación al archivo original
vinos["KMeans_Clusters"] = clustering.labels_ 
vinos
```

```python
#Visualizar los clusters que se formaron utiizando el metodo "PCA"
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_vinos = pca.fit_transform(vinos_norm)
pca_vinos_df = pd.DataFrame(data= pca_vinos, columns=["Componente_1", "Componente_2"])
pca_nombres_vinos = pd.concat([pca_vinos_df, vinos[["KMeans_Clusters"]]], axis=1)

pca_nombres_vinos
```

```python
fig = plt.figure(figsize= (6,6))

ax = fig.add_subplot(1,1,1)
ax.set_xlabel("Componente 1", fontsize=15)
ax.set_ylabel("Componente 2", fontsize=15)
ax.set_title("Componentes principales", fontsize=20)

color_theme = np.array(["blue", "green", "orange"])
ax.scatter(x = pca_nombres_vinos.Componente_1, y =pca_nombres_vinos.Componente_2,
            c = color_theme[pca_nombres_vinos.KMeans_Clusters], s = 50)
plt.show()
```
