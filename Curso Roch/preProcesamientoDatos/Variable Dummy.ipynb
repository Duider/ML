```python
import pandas as pd
```

```python
hidroponia = pd.read_csv("../arch/Hidroponia_original.csv", engine="python", index_col=0,encoding='latin-1')
hidroponia
```

```python
hidroponia.info()
```

```python
#Genero las variables dummy de hortaliza y sistema
hortaliza_dummy = pd.get_dummies(hidroponia["Hortaliza"], prefix="Hortaliza")
sistema_dummy = pd.get_dummies(hidroponia["Sistema"], prefix="Sistema")
```

```python
hortaliza_dummy
```

```python
sistema_dummy
```

```python
#unimos las variables de tipo dummy a la info original del archivo hidroponia
hidroponia = pd.concat([hidroponia, hortaliza_dummy], axis = 1)
hidroponia = pd.concat([hidroponia, sistema_dummy], axis = 1)
hidroponia
```

```python
#Eliminamos las columnas que no contienen variables tipo dummy
hidroponia_dummies = hidroponia.drop(["pH", "EC", "Hortaliza", "Sistema", "Duracion_Ciclo "],axis = 1)
#En vez de los nombre se puede poner el numero de columna
```
