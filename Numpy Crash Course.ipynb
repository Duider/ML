### Array indexing and Slicing

```python
#1-d
a = np.array([1,2,3,4])
a[-1]
```

```python
#2-d
a = np.array([[1,2,3,4],[5,6,7,8]])
print(a.shape)
print(a[-1][-1])
print(a[0][2])
```

```python
#3-d
a = np.array([[[1,2,3,4],[5,6,7,8]],
              [["a","b","c","d"],["e","f","g","h"]]])
print(a.shape)
print(a[1][1][1])
print(a[0][1][-1])
print("\n",a[1])
print("\n",a[1][1])
# El : ayuda bastante 
print("\n",a[1, :, 1:3])
```

### Where

```python
#is used as a filter condition
```

```python
a
```

```python
#Devuelve los indices en donde a es menor a 10
index = np.where(a<10)
a[index]
```

```python
#Esto es para ver que valores son menores que 10 y mayores que 5
b = a[index]
print(b)
index = np.where(b>5)
b[index]
```

### Concatenate and Sorting

```python
a = np.array([1,2,3,4])
b = np.array([3,4,5,6])
c = np.array([32,12,4,1])
```

```python
#Concatenar
x = np.concatenate([a,b,c])
x
```

```python
#Ordenar
x.sort()
x
```

```python
b = np.array([a, c])
b, b.ndim
```
