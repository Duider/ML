```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,4,3])
ax.text(x=2.5, y=2.5, s="ASDADSDAS", fontsize=20, color="red", bbox=dict(facecolor="coral", edgecolor="#003430", alpha=0.8,
                                                                         pad=1.3, boxstyle="circle"))

ax.annotate("Que onda",(2,2),xytext=(1,2), arrowprops=dict(width=8, headlength=30, headwidth=20, shrink=0.1, facecolor="green"))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,4,3],linestyle="--", label="dataset 1")
ax.plot([1,2,3,4],[4,5,6,4], label="dataset 2")
ax.scatter([1,2,3,4],[1.5,4,4.5,6], label="dataset 3")
ax.legend()#Muestra un cuadro con los labels de cada dataset
plt.show()
```
