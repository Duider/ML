```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [3,1,2,5,5,7,4,6]

fig , ax = plt.subplots()
ax.scatter(x,y, s=200, color="#ff0453", edgecolor=(0,0.5,0.3)) #"s" es el tamaño de los puntos 
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [3,1,2,5,5,7,4,6]
y2 = [1,3,2,5,5,7,4,6]
fig , ax = plt.subplots()
ax.plot(x,y, color="#0014f3",zorder=1) #"zorder" sirve para decidir que linea queda arriba de la otra, el mayor se vera por arriba
ax.plot(x,y2, color=(0.9,0.2,0.1),zorder=0) 
plt.show()
```

```python
x = ["A", "B", "C", "D", "E"]
y = [14,7,18,9,10]

fig, ax = plt.subplots()
ax.bar(x,y, color="grey", edgecolor="black")
plt.show()
```
