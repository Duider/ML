```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([[3,4,5,6,7,8],
            [3,4,5,6,7,8],
            [3,4,5,6,7,8],
            [1,3,5,7,8,9],
            [9,8,7,5,3,1],
            [3,4,5,6,7,8]])
fig, ax = plt.subplots()
a = ax.imshow(x, cmap=plt.cm.winter, interpolation="bilinear", extent= (1,8,2,10))
```

```python
r = np.linspace(0,2,20) # 20 valores del 0 al 2
#beta= np.pi*r
beta=2*np.pi*np.random.rand(20)
colors=beta
fig, ax = plt.subplots(subplot_kw=dict(projection="polar"))

sc = ax.scatter(beta,r, c=colors, s=100, cmap="hsv")
fig.colorbar(mappable=sc)
plt.show()
```
