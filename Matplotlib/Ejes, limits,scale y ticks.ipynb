```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import(MultipleLocator, AutoMinorLocator)
```

```python
x = np.linspace(0,8,200)
y= x**3
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_ylim(0,100)#Limites del eje y
ax.set_xticks((0,2,4,6,8),["Cero","Dos","Cuatro","Seis", "Ocho"])
#ax.set_xticks(np.arange(0,10,2),["Cero","Dos","Cuatro","Seis", "Ocho"]) Es lo mismo solo que por tramos

plt.show()
```

```python
x = np.linspace(0,4*np.pi,200)
y = 4+ np.cos(x)
fig, ax= plt.subplots()
ax.plot(x,y)
ax.set_ylim(0,10) #Sin este parametro no se verían los ejes desde su principio
plt.show()
```

```python
datos = {"Materias":["Ingles", "Matematica", "Lengua", "Historia","Física","Biología"], "Notas":[5.5,6,7,8.5,2,3]}
dfdatos=pd.DataFrame(datos) #Pasa los datos a DataFrame
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(dfdatos.Materias, dfdatos.Notas, color="#21A756") # Hace el grafico de barras
ax.grid(axis="y") #Genera lineas horizontales en los valores de "y"
plt.axhline(y = 7, color = '#0D5E42', linewidth=3) #Linea de aprobado
ax.set_yticks(range(0,11,1)) # que vaya del 0 al 10, en 1 en 1
ax.set_yticks(np.arange(0,11,0.5), minor=True)
#ax.yaxis.set_major_locator(MultipleLocator(1))
#ax.yaxis.set_minor_locator(MultipleLocator(0.5))
ax.yaxis.set_major_formatter("{x:.0f} pts")
plt.show()
```
