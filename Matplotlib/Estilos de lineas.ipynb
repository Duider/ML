```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.1,6)
y = np.sin(x)
plt.plot(x,y,linewidth=3,linestyle=(0,(2,3))) # (donde empieza la primera linea(espacio del trazo, espacio del hueco))
                                            #en el primer parametro suele ir siempre "0"
                                            #En la tupla se pueden poner mas de dos valores:
x = np.linspace(0.1,6)
y = np.cos(x)
plt.plot(x,y,linewidth=3,linestyle=(0,(8,1,3,5)), color="orange")        
plt.show()
```

```python
#Estilo stem
x = np.linspace(0.1,6)
y = np.sin(x)

fig , ax = plt.subplots()
ax.stem(x,y,linefmt=(":"), markerfmt="D", basefmt="C3-")
ax.spines["left"].set_color("red") #color
ax.spines["left"].set_linewidth(5) #ancho
ax.spines["left"].set_alpha(0.5) #transparencia
ax.spines["left"].set_linestyle("dashed")
ax.spines["bottom"].set_visible(0) #Borra la linea
ax.grid(True, linewidth=1)
plt.show()
```
