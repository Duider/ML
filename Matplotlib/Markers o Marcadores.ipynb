```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(6,6))
ax.plot([1,2,3,4,5,6],[5,4,7,6,8,6], marker="o",markerfacecolor="red", markersize=15, markeredgecolor="k", fillstyle="top",
        markerfacecoloralt="orange", markeredgewidth=1.5)
plt.show()
```

```python
import matplotlib
fig, ax = plt.subplots(figsize=(6,6))
vertice = matplotlib.path.Path(((0,0),(2,0),(2,2),(0,2)), codes=(1,2,2,2))
ax.plot([1,2,3,4,5,6],[5,4,7,6,8,6],marker=vertice,markersize=20)
plt.show()
```
