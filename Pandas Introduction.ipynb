```python
import pandas as pd
```

```python
#Series (one dimensional array)
flavors = ["mango", "cherry", "chocolate"]
print(pd.Series(flavors))

print("\n",pd.Series([1,2,3,4.0]))

print("\n",pd.Series([1,2,3, "four", "five", "six", True, False]))
```

```python
a = {"b": 2, "a": 1, "c": 3}
a = pd.Series(a)
print(pd.Series(a))
print(pd.Series(a, index=["a","b","c"]))
a.max(), a.min(), a.std(), a.var(), a.mean(), a.median(), a.count()
```

```python
s = pd.Series([1,2,3, "four", "five", "six", True, False])
s.count(), s.shape
```

### Pandas DataFrame Creation

```python
# A la función .DataFrame() hay que pasarle como parametro un diccionario. .T transpone la info
a = {"one": 1, "two": 2, "three": 3}
df = pd.DataFrame([a]).T
df
```

```python
a = ["one", "two", "three"]
pd.DataFrame(a, columns=["Numbers"])
```

```python
a = [["one", 1], ["two", 2,2.0 , "El Confiable"], ["three", 3 , 3.0]]
pd.DataFrame(a, columns=["String", "Int", "float", "Quiniela"])
```

```python
a = (("one", 1), ("two", 2,2.0 , "El Confiable"), ("three", 3 , 3.0))
pd.DataFrame(a, columns=["String", "Int", "float", "Quiniela"])
```
