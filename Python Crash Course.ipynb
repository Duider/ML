### Functions

```python
def my_function(parameters):
    print("Statements")
    print(parameters)
```

```python
my_function("PEPO DE LA VEGA")
```

```python
# el "*" antes del parametro significa que pueden variar el numero de parametros pasados para la función( en forma de tupla)
def add_numbers(*params):
    print("Sum of {} ==> {}".format(params, sum(params)))
    
add_numbers(1,2,3,4,5)
```

```python
# el "**" antes del parametro significa que pueden variar el numero de parametros pasados para la función (en forma de diccionario)
def add_numbers(**params):
    print(params, sum(params.values()))
    
    c = 0
    for x in params:
        c = c + params[x]
        
    print("sum: ",c)
    
    return c
    
add_numbers(a=1,b=2,c=3,d=4,e=5)
```
