```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
datos_cs = pd.read_csv("../arch/armasCs.csv")
```

```python
datos_cs
```

```python
datos_cs = datos_cs.rename(columns={"Pistols":"Weapons"})
datos_cs
```

```python
datos_cs.isnull().sum()
```

```python
datos_cs =datos_cs.dropna(thresh= 4)
datos_cs.reset_index(drop=True, inplace=True)
datos_cs
```

```python
pistols = datos_cs[0:13]
pistols
```

```python
shotguns = datos_cs[15:18]
shotguns
```

```python
SMG = datos_cs[19:26]
SMG
```

```python
rifles = datos_cs[27:38]
rifles
```

```python
LMG = datos_cs[39:41]
LMG
```

```python
snipers = datos_cs[42:-1]
snipers
```

```python
pistols["Armor\nPenetration"] = pistols["Armor\nPenetration"].str.rstrip("%")
pistols["Armor\nPenetration"] 
```

```python
mean_armorP_pistols = pistols["Armor\nPenetration"].drop([2,6,9]).astype(float)
mean_armorP_pistols = mean_armorP_pistols.mean()
mean_armorP_pistols
```

```python
shotguns["Armor\nPenetration"] = shotguns["Armor\nPenetration"].str.rstrip("%")
mean_armorP_shotguns = shotguns["Armor\nPenetration"].astype(float)
mean_armorP_shotguns = mean_armorP_shotguns.mean()
mean_armorP_shotguns = round(mean_armorP_shotguns,2)
mean_armorP_shotguns
```

```python
SMG["Armor\nPenetration"] = SMG["Armor\nPenetration"].str.rstrip("%")
mean_armorP_SMG = SMG["Armor\nPenetration"].astype(float)
mean_armorP_SMG = mean_armorP_SMG.mean()
mean_armorP_SMG = round(mean_armorP_SMG,2)
mean_armorP_SMG
```

```python
rifles["Armor\nPenetration"] = rifles["Armor\nPenetration"].str.rstrip("%")
mean_armorP_rifles = rifles["Armor\nPenetration"].drop([29,31,35,37]).astype(float)
mean_armorP_rifles = mean_armorP_rifles.mean()
mean_armorP_rifles = round(mean_armorP_rifles,2)
mean_armorP_rifles
```

```python
LMG["Armor\nPenetration"] = LMG["Armor\nPenetration"].str.rstrip("%")
mean_armorP_LMG = LMG["Armor\nPenetration"].astype(float)
mean_armorP_LMG = mean_armorP_LMG.mean()
mean_armorP_LMG
```

```python
snipers["Armor\nPenetration"] = snipers["Armor\nPenetration"].str.rstrip("%")
mean_armorP_snipers = snipers["Armor\nPenetration"].drop([43,45,47]).astype(float)
mean_armorP_snipers = mean_armorP_snipers.mean()
mean_armorP_snipers
```

```python
means_armorP={"mean_armorP_pistols":mean_armorP_pistols,
        "mean_armorP_shotguns": mean_armorP_shotguns,
        "mean_armorP_SMG": mean_armorP_SMG,
        "mean_armorP_rifles": mean_armorP_rifles,
        "mean_armorP_LMG": mean_armorP_LMG,
        "mean_armorP_snipers": mean_armorP_snipers}
means_armorP
```

```python
means_armorP.keys()
```

```python
means_armorP.values()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sc = ax.bar(means_armorP.keys(),means_armorP.values())
plt.ylim(0,100)
ax.text(x=-0.19,y=45, s="%"+str(means_armorP["mean_armorP_pistols"]),rotation=30, fontsize=13.5)
ax.text(x=0.79,y=40, s="%"+str(means_armorP["mean_armorP_shotguns"]),rotation=30, fontsize=13.5)
ax.text(x=1.82,y=35, s="%"+str(means_armorP["mean_armorP_SMG"]),rotation=30, fontsize=13.5)
ax.text(x=2.82,y=50, s="%"+str(means_armorP["mean_armorP_rifles"]),rotation=30, fontsize=13.5)
ax.text(x=3.87,y=47, s="%"+str(means_armorP["mean_armorP_LMG"]),rotation=30, fontsize=13.5)
ax.text(x=4.76,y=55, s="%"+str(means_armorP["mean_armorP_snipers"]),rotation=30, fontsize=13.5)
plt.show()



```

```python
mean_fireRate_pistols = pistols["Fire Rate\n(RPM)"].drop(9)
mean_fireRate_pistols[6] = mean_fireRate_pistols[6].replace(",","")
mean_fireRate_pistols = mean_fireRate_pistols.astype(float).mean().astype(int)
mean_fireRate_pistols 
```

```python
mean_fireRate_shotguns = shotguns["Fire Rate\n(RPM)"]
mean_fireRate_shotguns = mean_fireRate_shotguns.astype(float).mean().astype(int)
mean_fireRate_shotguns
```

```python
mean_fireRate_SMG = SMG["Fire Rate\n(RPM)"]
mean_fireRate_SMG = mean_fireRate_SMG.astype(float).mean().astype(int)
mean_fireRate_SMG
```

```python
mean_fireRate_rifles = rifles["Fire Rate\n(RPM)"]
mean_fireRate_rifles = mean_fireRate_rifles.drop([29,35,37])
mean_fireRate_rifles = mean_fireRate_rifles.astype(float).mean().astype(int)
mean_fireRate_rifles
```

```python
mean_fireRate_LMG = LMG["Fire Rate\n(RPM)"]
mean_fireRate_LMG = mean_fireRate_LMG.astype(float).mean().astype(int)
mean_fireRate_LMG
```

```python
mean_fireRate_snipers = snipers["Fire Rate\n(RPM)"]
mean_fireRate_snipers = mean_fireRate_snipers.drop([43,45,47])
mean_fireRate_snipers = mean_fireRate_snipers.astype(float).mean().astype(int)
mean_fireRate_snipers
```

```python
means_fireRate={"mean_fireRate_pistols":mean_fireRate_pistols,
        "mean_fireRate_shotguns": mean_fireRate_shotguns,
        "mean_fireRate_SMG": mean_fireRate_SMG,
        "mean_fireRate_rifles": mean_fireRate_rifles,
        "mean_fireRate_LMG": mean_fireRate_LMG,
        "mean_fireRate_snipers": mean_fireRate_snipers}
means_fireRate
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sc = ax.bar(means_fireRate.keys(),means_fireRate.values())
plt.ylim(0,1000)
ax.text(x=-0.1,y=200, s=str(means_fireRate["mean_fireRate_pistols"]),fontsize=13.5)
ax.text(x=0.9,y=30, s=str(means_fireRate["mean_fireRate_shotguns"]), fontsize=13.5)
ax.text(x=1.9,y=400, s=str(means_fireRate["mean_fireRate_SMG"]), fontsize=13.5)
ax.text(x=2.9,y=300, s=str(means_fireRate["mean_fireRate_rifles"]), fontsize=13.5)
ax.text(x=3.9,y=400, s=str(means_fireRate["mean_fireRate_LMG"]), fontsize=13.5)
ax.text(x=4.9,y=45, s=str(means_fireRate["mean_fireRate_snipers"]), fontsize=13.5)
plt.show()
```
